```python
#Data Cleaning and Preparation
```

```python
import pandas as pd
import numpy as np
```

```python
# Load the merged dataset
```

```python
df = pd.read_csv(r"K:\Instagram fake spammer project\merged_dataset.csv")
```

```python
df
```

```python
# Initial inspection
```

```python
print("Initial shape:", df.shape)
print("\nMissing values before cleaning:")
print(df.isnull().sum())
```

```python
# Handle missing values - fill with median for numerical, mode for categorical
```

```python
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
# Convert boolean columns to 0/1 (if any exist as strings)
```

```python
bool_cols = ['profile pic', 'private', 'name==username', 'external URL']
for col in bool_cols:
    if col in df.columns:
        df[col] = df[col].astype(int)
```

```python
# Remove exact duplicates
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Check for outliers in numerical columns
```

```python
num_cols = ['nums/length username', 'fullname words', 'nums/length fullname', 'description length', '#posts', '#followers', '#follows']
for col in num_cols:
    if col in df.columns:
        # Cap outliers at 99th percentile
        upper_limit = df[col].quantile(0.99)
        df[col] = np.where(df[col] > upper_limit, upper_limit, df[col])
```

```python
num_cols
```

```python
# Verify cleaning
```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
print("\nFinal shape:", df.shape)
print("\nData types:")
print(df.dtypes)
```

```python
# Save cleaned data
```

```python
df.to_csv('cleaned_instagram_data.csv', index=False)
```

```python
# Machine Learning Model Development
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
# Prepare data
X = df.drop('fake', axis=1)
y = df['fake']
```

```python
X
```

```python
y
```

```python
# Split data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale numerical features
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
scaler
```

```python
# Initialize and train Random Forest
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
# Predictions
```

```python
y_pred = rf.predict(X_test_scaled)
```

```python
y_pred
```

```python
# Evaluation
```

```python
print("Random Forest Performance:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nAccuracy:", accuracy_score(y_test, y_pred))
```

```python
# Feature Importance
```

```python
feature_importance = pd.DataFrame({'Feature': X.columns,'Importance': rf.feature_importances_}).sort_values('Importance', ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x='Importance', y='Feature', data=feature_importance.head(10))
plt.title('Top 10 Important Features for Fake Account Detection')
plt.show()
```

```python
# Hyperparameter tuning
```

```python
param_grid = {'n_estimators': [100, 200, 300],'max_depth': [None, 10, 20, 30],'min_samples_split': [2, 5, 10]}

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train_scaled, y_train)

best_rf = grid_search.best_estimator_
y_pred_best = best_rf.predict(X_test_scaled)

print("\nBest Model Performance:")
print(classification_report(y_test, y_pred_best))
print("\nBest Parameters:", grid_search.best_params_)
```
